## Customers Segmentation

Segmentation of customers can be used to define marketing strategies.


## Credit Card- Customer Segmentation Dataset

This dataset aims to develop a customer segmentation strategy for defining a marketing approach. It contains information on the usage behavior of around 9,000 active credit card holders over the past 6 months. The data is provided at the customer level and includes 18 behavioral variables.

### Data Dictionary

- **CUST_ID**: Identification of the credit card holder (Categorical)
- **BALANCE**: Remaining balance in the account for purchases
- **BALANCE_FREQUENCY**: Frequency of balance updates (score between 0 and 1; 1 = frequently updated, 0 = not frequently updated)
- **PURCHASES**: Total amount spent on purchases from the account
- **ONEOFF_PURCHASES**: Maximum purchase amount done in a single transaction
- **INSTALLMENTS_PURCHASES**: Total amount spent on purchases in installments
- **CASH_ADVANCE**: Amount of cash advance taken by the user
- **PURCHASES_FREQUENCY**: Frequency of purchases (score between 0 and 1; 1 = frequently purchased, 0 = not frequently purchased)
- **ONEOFFPURCHASESFREQUENCY**: Frequency of one-off purchases (1 = frequently, 0 = not frequently)
- **PURCHASESINSTALLMENTSFREQUENCY**: Frequency of installment purchases (1 = frequently, 0 = not frequently)
- **CASHADVANCEFREQUENCY**: Frequency of cash advances (1 = frequently, 0 = not frequently)
- **CASHADVANCETRX**: Number of transactions made with cash advances
- **PURCHASES_TRX**: Number of purchase transactions made
- **CREDIT_LIMIT**: Credit limit available to the user
- **PAYMENTS**: Total amount of payments made by the user
- **MINIMUM_PAYMENTS**: Minimum amount paid by the user
- **PRCFULLPAYMENT**: Percentage of full payment made by the use

```python
data= pd.read_csv("../input/CC GENERAL.csv")
print(data.shape)
data.head()
```

## Data Preprocessing

#### Descriptive Statistics of Data

#### Dealing with Missing Values

```python
data.isnull().sum().sort_values(ascending=False).head()
```

Imputing these missing values with mean

```python
data.loc[(data['MINIMUM_PAYMENTS'].isnull()==True),'MINIMUM_PAYMENTS']=data['MINIMUM_PAYMENTS'].mean()
data.loc[(data['CREDIT_LIMIT'].isnull()==True),'CREDIT_LIMIT']=data['CREDIT_LIMIT'].mean()
```

```python
data.isnull().sum().sort_values(ascending=False).head()
```


#### Dealing with Outliers

By dropping outliers we can lose many rows as there are too many outliers in dataset. So making ranges to deal with extreme values.

```python
columns=['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT',
        'PAYMENTS', 'MINIMUM_PAYMENTS']

for c in columns:
    
    Range=c+'_RANGE'
    data[Range]=0        
    data.loc[((data[c]>0)&(data[c]<=500)),Range]=1
    data.loc[((data[c]>500)&(data[c]<=1000)),Range]=2
    data.loc[((data[c]>1000)&(data[c]<=3000)),Range]=3
    data.loc[((data[c]>3000)&(data[c]<=5000)),Range]=4
    data.loc[((data[c]>5000)&(data[c]<=10000)),Range]=5
    data.loc[((data[c]>10000)),Range]=6
 
```

```python
columns=['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY', 
         'CASH_ADVANCE_FREQUENCY', 'PRC_FULL_PAYMENT']

for c in columns:
    
    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=0.1)),Range]=1
    data.loc[((data[c]>0.1)&(data[c]<=0.2)),Range]=2
    data.loc[((data[c]>0.2)&(data[c]<=0.3)),Range]=3
    data.loc[((data[c]>0.3)&(data[c]<=0.4)),Range]=4
    data.loc[((data[c]>0.4)&(data[c]<=0.5)),Range]=5
    data.loc[((data[c]>0.5)&(data[c]<=0.6)),Range]=6
    data.loc[((data[c]>0.6)&(data[c]<=0.7)),Range]=7
    data.loc[((data[c]>0.7)&(data[c]<=0.8)),Range]=8
    data.loc[((data[c]>0.8)&(data[c]<=0.9)),Range]=9
    data.loc[((data[c]>0.9)&(data[c]<=1.0)),Range]=10
    
```

```python
columns=['PURCHASES_TRX', 'CASH_ADVANCE_TRX']  

for c in columns:
    
    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=5)),Range]=1
    data.loc[((data[c]>5)&(data[c]<=10)),Range]=2
    data.loc[((data[c]>10)&(data[c]<=15)),Range]=3
    data.loc[((data[c]>15)&(data[c]<=20)),Range]=4
    data.loc[((data[c]>20)&(data[c]<=30)),Range]=5
    data.loc[((data[c]>30)&(data[c]<=50)),Range]=6
    data.loc[((data[c]>50)&(data[c]<=100)),Range]=7
    data.loc[((data[c]>100)),Range]=8
```

```python
data.drop(['CUST_ID', 'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES',
       'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE',
       'PURCHASES_FREQUENCY',  'ONEOFF_PURCHASES_FREQUENCY',
       'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY',
       'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT' ], axis=1, inplace=True)

X= np.asarray(data)
```

Normalizing input values.

```python
scale = StandardScaler()
X = scale.fit_transform(X)
X.shape
```

## MODELING

#### Clustering using Kmeans

```python
n_clusters=30
cost=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(X)
    cost.append(kmean.inertia_)  
```

```python
plt.plot(cost, 'bx-', color = "red")
```

### Silhuete score

```python
from sklearn.metrics import silhouette_score

silhouette = []
for k in range(2,11):
    kmeans= KMeans(i)
    kmeans.fit(data)
    S_score = silhouette_score(data, kmeans.labels_)
    silhouette.append(S_score)

plt.plot(silhouette, "-o", color = "black")
```

#### Choosing 6 no of clusters 

```python
kmean= KMeans(6)
kmean.fit(X)
labels=kmean.labels_
```

```python
clusters=pd.concat([data, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head()
```

## Interpretation of Clusters

```python
for c in clusters:
    grid= sns.FacetGrid(clusters, col='cluster')
    grid.map(plt.hist, c)
```
