## presentación de arquitectura ARLSTM





### Resumen de conceptos fundamentales de las ARLSTM (Autoregressive LSTM)

1. **Definición de ARLSTM**:  
   Las ARLSTM son modelos de aprendizaje profundo basados en redes LSTM (Long Short-Term Memory) diseñados para la predicción de series temporales. A diferencia de los modelos multi-step tradicionales, que generan toda la secuencia de predicciones en una sola operación, las ARLSTM generan predicciones de manera iterativa y autoregresiva. Esto significa que cada predicción se retroalimenta en el modelo para generar la siguiente, creando pronósticos progresivos.

2. **Ventajas de las ARLSTM**:
   - **Flexibilidad en la longitud de la secuencia**: Permiten generar secuencias de cualquier longitud sin necesidad de reentrenar el modelo. Esto simplifica la generación de pronósticos a diferentes escalas temporales (horas, días, meses, etc.).
   - **Adaptabilidad**: Esta arquitectura se adapta bien a tareas donde la longitud de la predicción puede variar, evitando la necesidad de modificar ventanas de datos o estructuras del modelo.

3. **Aplicaciones destacadas**:
   - **WaveNet**: Un modelo de Google DeepMind que utiliza esta arquitectura para generar secuencias de audio crudo.
   - **DeepAR**: Una metodología de Amazon que emplea modelos autoregresivos recurrentes para lograr resultados de vanguardia en series temporales.

4. **Caveat: Acumulación de errores**:
   - Debido a que las predicciones se utilizan como entrada para generar las siguientes, los errores presentes en predicciones iniciales pueden acumularse progresivamente. Esto puede llevar a que las predicciones finales tengan un margen de error más alto, afectando la precisión general del modelo.

5. **Importancia de protocolos rigurosos de prueba**:  
   Es fundamental evaluar cuidadosamente las ARLSTM para determinar si son la solución más adecuada para un problema específico, especialmente considerando su sensibilidad al error acumulativo.

6. **Implementación práctica**:  
   Las ARLSTM se pueden construir fácilmente en frameworks como Keras. En este caso, se comparan con modelos multi-step tradicionales para evaluar su desempeño en series temporales.

7. **Uso típico en series temporales**:  
   Las ARLSTM son ideales para problemas que requieren predicciones continuas o escalables y ofrecen una solución potente para manejar dinámicas temporales complejas.

### Punto clave:  
Aunque son modelos flexibles y escalables, las ARLSTM deben manejarse con cuidado debido a la acumulación de errores, lo que resalta la necesidad de realizar pruebas exhaustivas para garantizar resultados fiables.

### En terminos intuitivos:

Las ARLSTM son un tipo de modelo de redes LSTM que predicen paso a paso en lugar de generar toda la secuencia de una vez. Usan su propia predicción previa como entrada para calcular la siguiente, lo que permite crear pronósticos de cualquier longitud sin tener que reentrenar el modelo.

Esto es útil para series temporales con diferentes escalas (horas, días, meses) y se aplica en modelos avanzados como WaveNet o DeepAR. Sin embargo, tienen un problema: si las primeras predicciones tienen errores, estos se acumulan y empeoran las predicciones siguientes. Por eso, es clave probarlas bien antes de usarlas en un proyecto.

```python
class DataWindow():
    def __init__(self, input_width, label_width, shift, 
                 train_df=train_df, val_df=val_df, test_df=test_df, 
                 label_columns=None):
        
        self.train_df = train_df
        self.val_df = val_df
        self.test_df = test_df
        
        self.label_columns = label_columns
        if label_columns is not None:
            self.label_columns_indices = {name: i for i, name in enumerate(label_columns)}
        self.column_indices = {name: i for i, name in enumerate(train_df.columns)}
        
        self.input_width = input_width
        self.label_width = label_width
        self.shift = shift
        
        self.total_window_size = input_width + shift
        
        self.input_slice = slice(0, input_width)
        self.input_indices = np.arange(self.total_window_size)[self.input_slice]
        
        self.label_start = self.total_window_size - self.label_width
        self.labels_slice = slice(self.label_start, None)
        self.label_indices = np.arange(self.total_window_size)[self.labels_slice]
    
    def split_to_inputs_labels(self, features):
        inputs = features[:, self.input_slice, :]
        labels = features[:, self.labels_slice, :]
        if self.label_columns is not None:
            labels = tf.stack(
                [labels[:,:,self.column_indices[name]] for name in self.label_columns],
                axis=-1
            )
        inputs.set_shape([None, self.input_width, None])
        labels.set_shape([None, self.label_width, None])
        
        return inputs, labels
    
    def plot(self, model=None, plot_col='traffic_volume', max_subplots=3):
        inputs, labels = self.sample_batch
        
        plt.figure(figsize=(12, 8))
        plot_col_index = self.column_indices[plot_col]
        max_n = min(max_subplots, len(inputs))
        
        for n in range(max_n):
            plt.subplot(3, 1, n+1)
            plt.ylabel(f'{plot_col} [scaled]')
            plt.plot(self.input_indices, inputs[n, :, plot_col_index],
                     label='Inputs', marker='.', zorder=-10)

            if self.label_columns:
              label_col_index = self.label_columns_indices.get(plot_col, None)
            else:
              label_col_index = plot_col_index

            if label_col_index is None:
              continue

            plt.scatter(self.label_indices, labels[n, :, label_col_index],
                        edgecolors='k', marker='s', label='Labels', c='green', s=64)
            if model is not None:
              predictions = model(inputs)
              plt.scatter(self.label_indices, predictions[n, :, label_col_index],
                          marker='X', edgecolors='k', label='Predictions',
                          c='red', s=64)

            if n == 0:
              plt.legend()

        plt.xlabel('Time (h)')
        
    def make_dataset(self, data):
        data = np.array(data, dtype=np.float32)
        ds = tf.keras.preprocessing.timeseries_dataset_from_array(
            data=data,
            targets=None,
            sequence_length=self.total_window_size,
            sequence_stride=1,
            shuffle=True,
            batch_size=32
        )
        
        ds = ds.map(self.split_to_inputs_labels)
        return ds
    
    @property
    def train(self):
        return self.make_dataset(self.train_df)
    
    @property
    def val(self):
        return self.make_dataset(self.val_df)
    
    @property
    def test(self):
        return self.make_dataset(self.test_df)
    
    @property
    def sample_batch(self):
        result = getattr(self, '_sample_batch', None)
        if result is None:
            result = next(iter(self.train))
            self._sample_batch = result
        return result
```

```python
def compile_and_fit(model, window, patience=3, max_epochs=50):
    early_stopping = EarlyStopping(monitor='val_loss',
                                   patience=patience,
                                   mode='min')
    
    model.compile(loss=MeanSquaredError(),
                  optimizer=Adam(),
                  metrics=[MeanAbsoluteError()])
    
    history = model.fit(window.train,
                       epochs=max_epochs,
                       validation_data=window.val,
                       callbacks=[early_stopping])
    
    return history
```

## Autoregressive LSTM 



```python
multi_window = DataWindow(input_width=24, label_width=24, shift=24, label_columns=['traffic_volume'])
```

```python
# Values obtained from chapter 16

ms_mae_val = [0.352, 0.347, 0.088, 0.078, 0.070, 0.078, 0.069]
ms_mae_test = [0.347, 0.341, 0.076, 0.064, 0.058, 0.063, 0.055]
```

```python
class AutoRegressive(Model):
    def __init__(self, units, out_steps):
        super().__init__()
        self.out_steps = out_steps
        self.units = units
        self.lstm_cell = LSTMCell(units)
        self.lstm_rnn = RNN(self.lstm_cell, return_state=True)
        self.dense = Dense(train_df.shape[1])
        
    def warmup(self, inputs):
        x, *state = self.lstm_rnn(inputs)
        prediction = self.dense(x)
                
        return prediction, state
    
    def call(self, inputs, training=None):
        predictions = []
        prediction, state = self.warmup(inputs)
        
        predictions.append(prediction)
        
        for n in range(1, self.out_steps):
            x = prediction
            x, state = self.lstm_cell(x, states=state, training=training)
            
            prediction = self.dense(x)
            predictions.append(prediction)
        
        predictions = tf.stack(predictions)
        predictions = tf.transpose(predictions, [1, 0, 2])
        
        return predictions
```

```python
AR_LSTM = AutoRegressive(units=32, out_steps=24)

history = compile_and_fit(AR_LSTM, multi_window)

ms_val_performance = {}
ms_performance = {}

ms_val_performance['AR - LSTM'] = AR_LSTM.evaluate(multi_window.val)
ms_performance['AR - LSTM'] = AR_LSTM.evaluate(multi_window.test, verbose=0)
```
