En modelos de aprendizaje profundo para series temporales, no utilizamos funciones predeterminadas, sino que la red neuronal aprende a generar su propia funcin para hacer las mejores predicciones posibles. Este proceso se llama **entrenamiento**. Consiste en tomar una secuencia de datos, donde definimos qu puntos sern las entradas (valores anteriores) y cules las salidas (predicciones deseadas). El modelo ajusta sus parmetros, realiza predicciones, las compara con las salidas reales y repite el proceso hasta mejorar la precisin.

Por ejemplo, utilizamos una **ventana de datos** de 24 horas para predecir las siguientes 24 horas. Aunque el aprendizaje profundo maneja grandes volmenes de datos, la clave est en cmo dividimos los datos en ventanas. Cada ventana usa 24 horas de datos para generar una prediccin de las siguientes 24 horas. Al dividir todo el conjunto de datos en varias ventanas, obtenemos muchas secuencias con entradas y salidas que el modelo puede usar para entrenar y mejorar sus predicciones.

# Implementacin de la clase DataWindow

 Esta clase tiene 
la ventaja de ser flexible, lo que significa que puede usarse en una amplia 
variedad de escenarios para aplicar el aprendizaje profundo

La clase se basa en el ancho de la entrada, el ancho de la etiqueta y el cambio. El 
ancho de la entrada es simplemente la cantidad de intervalos de tiempo que 
se introducen en el modelo para realizar predicciones. Por ejemplo, dado que 
tenemos datos por hora en nuestro conjunto de datos, si introducimos en el 
modelo 24 horas de datos para realizar una prediccin, el ancho de entrada es 
24. Si introducimos solo 12 horas de datos, el ancho de entrada es 12

El ancho de la etiqueta es equivalente a la cantidad de intervalos de tiempo en 
las predicciones. Si predecimos solo un intervalo de tiempo, el ancho de la 
etiqueta es 1. Si predecimos un da completo de datos (con datos por hora), el 
ancho de la etiqueta es 24

Por ltimo, el desplazamiento es el nmero de intervalos de tiempo que separan 
la entrada y las predicciones. Si predecimos el siguiente intervalo de tiempo, el 
desplazamiento es 1. Si predecimos las siguientes 24 horas (con datos 
horarios), el desplazamiento es 24

Visualicemos algunas ventanas de datos para comprender mejor estos parametros

La clase `DataWindow` se utiliza para preparar y organizar los datos de series temporales en "ventanas" que se usarn para entrenar modelos de aprendizaje profundo.

# Explicacin de la DataWindow Class:

### 1. **Parmetros y Inicializacin (`__init__`)**
   Al crear una instancia de la clase `DataWindow`, se definen varios parmetros importantes:

   - **`input_width`**: Nmero de intervalos de tiempo que usas para hacer la prediccin. Por ejemplo, si usas 24 horas de datos para predecir, este valor ser 24.
   - **`label_width`**: Nmero de intervalos de tiempo que deseas predecir. Por ejemplo, si deseas predecir las prximas 24 horas, este valor ser 24.
   - **`shift`**: Desplazamiento entre las entradas y las etiquetas. Si predices el siguiente intervalo de tiempo (es decir, el siguiente dato), este valor ser 1.

   **Ventajas de estos parmetros**:
   - La clase permite elegir cuntas horas de datos usas como entrada (input) y cuntas horas deseas predecir (output).
   - El desplazamiento (`shift`) ayuda a manejar la separacin entre las predicciones y las entradas.

### 2. **Organizacin de Ventanas de Datos**
   Con los parmetros dados, la clase organiza los datos en ventanas de tamao `input_width + shift`. Por ejemplo, si `input_width = 24` y `shift = 1`, la ventana total de datos tiene 25 pasos de tiempo.

   La entrada (`inputs`) se toma desde el principio de la ventana hasta `input_width`, y las etiquetas (`labels`) se toman desde el final de la ventana hacia atrs, con un tamao de `label_width`.

   As, si tienes datos por hora y quieres predecir 24 horas basndote en las ltimas 24 horas, tendrs 48 puntos en total: 24 para las entradas y 24 para las etiquetas.

### 3. **Transformacin de los Datos (`split_to_inputs_labels`)**
   Este mtodo toma los datos y los divide en entradas y etiquetas:
   - **Entradas (`inputs`)**: Son los valores de los primeros `input_width` pasos de tiempo.
   - **Etiquetas (`labels`)**: Son los valores de los ltimos `label_width` pasos de tiempo.
   
   Si se especifican columnas de etiquetas, como el trfico de vehculos, el mtodo selecciona solo esas columnas para las etiquetas.

### 4. **Generacin de los Datos para el Modelo (`make_dataset`)**
   Este mtodo convierte los datos en un formato que puede ser utilizado para el entrenamiento del modelo. Utiliza la funcin `timeseries_dataset_from_array` para crear un conjunto de datos a partir de las ventanas de datos generadas. Los datos se dividen en lotes (batches) y se reorganizan en secuencias para el entrenamiento.

### 5. **Visualizacin de las Ventanas de Datos (`plot`)**
   El mtodo `plot` permite visualizar cmo se ven las entradas y etiquetas para un parmetro especfico (por ejemplo, el volumen de trfico). Las entradas se muestran con puntos y las etiquetas con cruces. Si se proporciona un modelo, tambin puede mostrar las predicciones del modelo en el grfico.

### 6. **Propiedades de los Conjuntos de Datos (`train`, `val`, `test`)**
   Las propiedades `train`, `val`, y `test` devuelven los conjuntos de datos de entrenamiento, validacin y prueba, respectivamente, utilizando el mtodo `make_dataset` para procesar los datos en estos conjuntos.

### 7. **Generacin de un Lote de Datos (`sample_batch`)**
   El mtodo `sample_batch` proporciona un lote de datos de entrenamiento para visualizar cmo se estructuran las ventanas de datos.

```python
class DataWindow():
    def __init__(self, input_width, label_width, shift, 
                 train_df=train_df, val_df=val_df, test_df=test_df, 
                 label_columns=None):
        
        self.train_df = train_df
        self.val_df = val_df
        self.test_df = test_df
        
        self.label_columns = label_columns
        if label_columns is not None:
            self.label_columns_indices = {name: i for i, name in enumerate(label_columns)}
        self.column_indices = {name: i for i, name in enumerate(train_df.columns)}
        
        self.input_width = input_width
        self.label_width = label_width
        self.shift = shift
        
        self.total_window_size = input_width + shift
        
        self.input_slice = slice(0, input_width)
        self.input_indices = np.arange(self.total_window_size)[self.input_slice]
        
        self.label_start = self.total_window_size - self.label_width
        self.labels_slice = slice(self.label_start, None)
        self.label_indices = np.arange(self.total_window_size)[self.labels_slice]

    #  La funcin split_to_inputs_labels simplemente separa la ventana de  big data en dos ventanas, donde una contiene las entradas y la otra las  etiquetas
    def split_to_inputs_labels(self, features):
        inputs = features[:, self.input_slice, :]
        labels = features[:, self.labels_slice, :]
        if self.label_columns is not None:
            labels = tf.stack(
                [labels[:,:,self.column_indices[name]] for name in self.label_columns],
                axis=-1
            )
        inputs.set_shape([None, self.input_width, None])
        labels.set_shape([None, self.label_width, None])
        
        return inputs, labels

    # Mtodo para representar grficamente una muestra de ventanas de datos
    def plot(self, model=None, plot_col='traffic_volume', max_subplots=3):
        inputs, labels = self.sample_batch
        
        plt.figure(figsize=(12, 8))
        plot_col_index = self.column_indices[plot_col]
        max_n = min(max_subplots, len(inputs))
        
        for n in range(max_n):
            plt.subplot(3, 1, n+1)
            plt.ylabel(f'{plot_col} [scaled]')
            plt.plot(self.input_indices, inputs[n, :, plot_col_index],
                     label='Inputs', marker='.', zorder=-10)

            if self.label_columns:
              label_col_index = self.label_columns_indices.get(plot_col, None)
            else:
              label_col_index = plot_col_index

            if label_col_index is None:
              continue

            plt.scatter(self.label_indices, labels[n, :, label_col_index],
                        edgecolors='k', marker='s', label='Labels', c='green', s=64)
            if model is not None:
              predictions = model(inputs)
              plt.scatter(self.label_indices, predictions[n, :, label_col_index],
                          marker='X', edgecolors='k', label='Predictions',
                          c='red', s=64)

            if n == 0:
              plt.legend()

        plt.xlabel('Time (h)')
        
    def make_dataset(self, data):
        data = np.array(data, dtype=np.float32)
        ds = tf.keras.preprocessing.timeseries_dataset_from_array(
            data=data,
            targets=None,
            sequence_length=self.total_window_size,
            sequence_stride=1,
            shuffle=True,
            batch_size=32
        )
        
        ds = ds.map(self.split_to_inputs_labels)
        return ds
    
    @property
    def train(self):
        return self.make_dataset(self.train_df)
    
    @property
    def val(self):
        return self.make_dataset(self.val_df)
    
    @property
    def test(self):
        return self.make_dataset(self.test_df)
    
    @property
    def sample_batch(self):
        result = getattr(self, '_sample_batch', None)
        if result is None:
            result = next(iter(self.train))
            self._sample_batch = result
        return result
```

## 13.2 Applying baseline models
### 13.2.1 Single-step baseline model

```python
single_step_window = DataWindow(input_width=1, label_width=1, shift=1, label_columns=['traffic_volume']) 
wide_window = DataWindow(input_width=24, label_width=24, shift=1, label_columns=['traffic_volume'])
```

```python
class Baseline(Model):
    def __init__(self, label_index=None):
        super().__init__()
        self.label_index = label_index
        
    def call(self, inputs):
        if self.label_index is None:
            return inputs
        
        elif isinstance(self.label_index, list):
            tensors = []
            for index in self.label_index:
                result = inputs[:, :, index]
                result = result[:, :, tf.newaxis]
                tensors.append(result)
            return tf.concat(tensors, axis=-1)
        
        result = inputs[:, :, self.label_index]
        return result[:,:,tf.newaxis]
```

### 13.2.2 Multi-step baseline models 

```python
multi_window = DataWindow(input_width=24, label_width=24, shift=24, label_columns=['traffic_volume'])
```

#### Predicting the last known value 

```python
class MultiStepLastBaseline(Model):
    def __init__(self, label_index=None):
        super().__init__()
        self.label_index = label_index
        
    def call(self, inputs):
        if self.label_index is None:
            return tf.tile(inputs[:, -1:, :], [1, 24, 1])
        return tf.tile(inputs[:, -1:, self.label_index:], [1, 24, 1])
```

```python
ms_baseline_last = MultiStepLastBaseline(label_index=column_indices['traffic_volume'])

ms_baseline_last.compile(loss=MeanSquaredError(), metrics=[MeanAbsoluteError()])

ms_val_performance = {}
ms_performance = {}

ms_val_performance['Baseline - Last'] = ms_baseline_last.evaluate(multi_window.val)
ms_performance['Baseline - Last'] = ms_baseline_last.evaluate(multi_window.test, verbose=0)
```

#### Repeating the input sequence 

Implementemos una segunda lnea base para los modelos de varios pasos, 
que simplemente devuelve la secuencia de entrada. Esto significa que la 
prediccin para las prximas 24 horas ser simplemente la ltima de las 24 
horas de datos conocidas

```python
class RepeatBaseline(Model):
    def __init__(self, label_index=None):
        super().__init__()
        self.label_index = label_index
        
    def call(self, inputs):
        return inputs[:, :, self.label_index:]
```

```python
ms_baseline_repeat = RepeatBaseline(label_index=column_indices['traffic_volume'])

ms_baseline_repeat.compile(loss=MeanSquaredError(), metrics=[MeanAbsoluteError()])

ms_val_performance['Baseline - Repeat'] = ms_baseline_repeat.evaluate(multi_window.val)
ms_performance['Baseline - Repeat'] = ms_baseline_repeat.evaluate(multi_window.test, verbose=0)
```

A diferencia de un modelo base, un modelo lineal requiere entrenamiento. Por lo tanto, definiremos una funcin compile_and_fit que configura el modelo para el entrenamiento y luego ajusta el modelo a los datos.

## 14.1 Implementing a linear model
### 14.1.1 Building a single-step linear model

```python
def compile_and_fit(model, window, patience=3, max_epochs=50):
    early_stopping = EarlyStopping(monitor='val_loss',
                                   patience=patience,
                                   mode='min')
    
    model.compile(loss=MeanSquaredError(),
                  optimizer=Adam(),
                  metrics=[MeanAbsoluteError()])
    
    history = model.fit(window.train,
                       epochs=max_epochs,
                       validation_data=window.val,
                       callbacks=[early_stopping])
    
    return history
```

```python
linear = Sequential([
    Dense(units=1)
])

history = compile_and_fit(linear, single_step_window)

val_performance['Linear'] = linear.evaluate(single_step_window.val)
performance['Linear'] = linear.evaluate(single_step_window.test, verbose=0)
```
