### Bibliotecas para la prediccin automtica
Existen varias bibliotecas que automatizan el proceso de prediccin de series temporales, entre las ms populares se encuentran:
- **Pmdarima**: implementacin de la biblioteca auto.arima de R.
- **Prophet**: desarrollada por Meta, es ampliamente utilizada para predicciones empresariales a gran escala, destacando por su capacidad para modelar tendencias no lineales y efectos estacionales mltiples.
- **NeuralProphet**: extensin de Prophet que integra modelos hbridos, como ARIMA y redes neuronales.
- **PyTorch Forecasting**: biblioteca que utiliza modelos avanzados de aprendizaje profundo para prediccin de series temporales, como DeepAR y N-Beats.

Cada una de estas bibliotecas tiene ventajas y limitaciones dependiendo del contexto, pero en este captulo nos enfocamos en **Prophet** debido a su popularidad y versatilidad.

### Qu es Prophet?
**Prophet** es una biblioteca de cdigo abierto creada por Meta, diseada para predicciones rpidas y eficientes de series temporales. Est especialmente enfocada en modelar series temporales con mltiples perodos estacionales y tendencias no lineales. Prophet utiliza un modelo aditivo general en el que una serie temporal \( y(t) \) se modela como una combinacin lineal de varios componentes: tendencia (\( g(t) \)), estacionalidad (\( s(t) \)), efectos de das festivos (\( h(t) \)) y un trmino de error \( \epsilon_t \).

### Componentes del modelo de Prophet:
1. **Tendencia**: modela los cambios no peridicos a largo plazo.
2. **Estacionalidad**: modela los cambios peridicos, como los efectos estacionales anuales, mensuales, semanales o diarios.
3. **Efectos de das festivos**: permite incorporar el impacto de das festivos especficos en la serie temporal.
4. **Error**: captura los cambios no explicados por los otros componentes.

La inclusin de efectos estacionales se realiza mediante series de Fourier, lo que permite modelar mltiples perodos estacionales simultneamente. Prophet es robusto frente a valores atpicos y datos faltantes, lo que lo hace ideal para aplicaciones comerciales.

### Consideraciones y flexibilidad de Prophet:
Prophet es flexible y permite ajustes finos para mejorar la precisin de las predicciones, pero no es una solucin nica para todos los casos. Funciona mejor con series temporales que tienen un fuerte componente estacional y varios ciclos histricos de datos. Aunque es muy til para ciertos tipos de series temporales, en otros casos podra no ser la mejor opcin, pero otras herramientas estadsticas y de aprendizaje profundo pueden ofrecer alternativas efectivas.

El captulo tambin hace referencia a recursos adicionales para profundizar en la comprensin de Prophet, como el artculo oficial "Forecasting at Scale" de Sean J. Taylor y Benjamin Letham.

En resumen, **Prophet** es una herramienta poderosa para la prediccin rpida de series temporales, ideal para quienes necesitan manejar datos con estacionalidad y tendencias no lineales.

### 19.4.3 Hyperparameter tuning 

```python
from itertools import product

param_grid = {
    'changepoint_prior_scale': [0.001, 0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 10.0]
}

all_params = [dict(zip(param_grid.keys(), v)) for v in product(*param_grid.values())]

maes = []

for params in all_params:
    m = Prophet(**params).fit(train)
    df_cv = cross_validation(m, initial='730 days', period='180 days', horizon='365 days', parallel='processes')
    df_p = performance_metrics(df_cv, rolling_window=1)
    maes.append(df_p['mae'].values[0])
    
tuning_results = pd.DataFrame(all_params)
tuning_results['mae'] = maes

tuning_results
```

```python
best_params = all_params[np.argmin(maes)]
print(best_params)
```

### 19.5.2 Experiment: Can SARIMA do better? 

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from tqdm import tqdm_notebook
from itertools import product
from typing import Union
```

```python
ad_fuller_result = adfuller(df['y'])

print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```

```python
y_diff = np.diff(df['y'], n=1)

ad_fuller_result = adfuller(y_diff)

print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```

$d=1$, $D=0$ and $m=12$

```python
def optimize_SARIMAX(endog: Union[pd.Series, list], exog: Union[pd.Series, list], order_list: list, d: int, D: int, s: int) -> pd.DataFrame:
    
    results = []
    
    for order in tqdm_notebook(order_list):
        try: 
            model = SARIMAX(
                endog,
                exog,
                order=(order[0], d, order[1]),
                seasonal_order=(order[2], D, order[3], s),
                simple_differencing=False).fit(disp=False)
        except:
            continue
            
        aic = model.aic
        results.append([order, model.aic])
        
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q,P,Q)', 'AIC']
    
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)
    
    return result_df
```

```python
ps = range(0, 4, 1)
qs = range(0, 4, 1)
Ps = range(0, 4, 1)
Qs = range(0, 4, 1)

order_list = list(product(ps, qs, Ps, Qs))

d = 1
D = 0
s = 12
```

```python
SARIMA_result_df = optimize_SARIMAX(train['y'], None, order_list, d, D, s)
SARIMA_result_df
```

```python
SARIMA_model = SARIMAX(train['y'], order=(1,1,1), seasonal_order=(1,0,1,12), simple_differencing=False)
SARIMA_model_fit = SARIMA_model.fit(disp=False)

print(SARIMA_model_fit.summary())
```

```python
SARIMA_model_fit.plot_diagnostics(figsize=(10,8));
```

```python
residuals = SARIMA_model_fit.resid

lbvalue, pvalue = acorr_ljungbox(residuals, np.arange(1, 11, 1))

print(pvalue)
```
