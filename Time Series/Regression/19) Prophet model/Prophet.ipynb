
### Resumen del Capítulo 19

**Automatización de la predicción de series temporales con Prophet**
Este capítulo analiza cómo automatizar la predicción de series temporales utilizando bibliotecas específicas, destacando el uso de **Prophet**, una de las herramientas más conocidas y utilizadas en la industria.

A lo largo del libro, hemos construido modelos que requieren varios pasos manuales, como la selección del mejor modelo en SARIMAX o la creación de clases para manejar redes neuronales en la parte de aprendizaje profundo. Si bien esto ofrece gran flexibilidad y control total sobre los modelos, también es útil automatizar el proceso de predicción para facilitar la predicción de series temporales y acelerar los experimentos.

### Bibliotecas para la predicción automática
Existen varias bibliotecas que automatizan el proceso de predicción de series temporales, entre las más populares se encuentran:
- **Pmdarima**: implementación de la biblioteca auto.arima de R.
- **Prophet**: desarrollada por Meta, es ampliamente utilizada para predicciones empresariales a gran escala, destacando por su capacidad para modelar tendencias no lineales y efectos estacionales múltiples.
- **NeuralProphet**: extensión de Prophet que integra modelos híbridos, como ARIMA y redes neuronales.
- **PyTorch Forecasting**: biblioteca que utiliza modelos avanzados de aprendizaje profundo para predicción de series temporales, como DeepAR y N-Beats.

Cada una de estas bibliotecas tiene ventajas y limitaciones dependiendo del contexto, pero en este capítulo nos enfocamos en **Prophet** debido a su popularidad y versatilidad.

### ¿Qué es Prophet?
**Prophet** es una biblioteca de código abierto creada por Meta, diseñada para predicciones rápidas y eficientes de series temporales. Está especialmente enfocada en modelar series temporales con múltiples períodos estacionales y tendencias no lineales. Prophet utiliza un modelo aditivo general en el que una serie temporal \( y(t) \) se modela como una combinación lineal de varios componentes: tendencia (\( g(t) \)), estacionalidad (\( s(t) \)), efectos de días festivos (\( h(t) \)) y un término de error \( \epsilon_t \).

### Componentes del modelo de Prophet:
1. **Tendencia**: modela los cambios no periódicos a largo plazo.
2. **Estacionalidad**: modela los cambios periódicos, como los efectos estacionales anuales, mensuales, semanales o diarios.
3. **Efectos de días festivos**: permite incorporar el impacto de días festivos específicos en la serie temporal.
4. **Error**: captura los cambios no explicados por los otros componentes.

La inclusión de efectos estacionales se realiza mediante series de Fourier, lo que permite modelar múltiples períodos estacionales simultáneamente. Prophet es robusto frente a valores atípicos y datos faltantes, lo que lo hace ideal para aplicaciones comerciales.

### Consideraciones y flexibilidad de Prophet:
Prophet es flexible y permite ajustes finos para mejorar la precisión de las predicciones, pero no es una solución única para todos los casos. Funciona mejor con series temporales que tienen un fuerte componente estacional y varios ciclos históricos de datos. Aunque es muy útil para ciertos tipos de series temporales, en otros casos podría no ser la mejor opción, pero otras herramientas estadísticas y de aprendizaje profundo pueden ofrecer alternativas efectivas.

El capítulo también hace referencia a recursos adicionales para profundizar en la comprensión de Prophet, como el artículo oficial "Forecasting at Scale" de Sean J. Taylor y Benjamin Letham.

En resumen, **Prophet** es una herramienta poderosa para la predicción rápida de series temporales, ideal para quienes necesitan manejar datos con estacionalidad y tendencias no lineales.

## Basic forecasting with Prophet

```python
df.columns = ['ds', 'y']
df.head()
```

```python
m.fit(train);
```

```python
future = m.make_future_dataframe(periods=365)
```

```python
forecast = m.predict(future)

forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].iloc[-365:-360]
```

### 19.4.3 Hyperparameter tuning 

```python
from itertools import product

param_grid = {
    'changepoint_prior_scale': [0.001, 0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 10.0]
}

all_params = [dict(zip(param_grid.keys(), v)) for v in product(*param_grid.values())]

maes = []

for params in all_params:
    m = Prophet(**params).fit(train)
    df_cv = cross_validation(m, initial='730 days', period='180 days', horizon='365 days', parallel='processes')
    df_p = performance_metrics(df_cv, rolling_window=1)
    maes.append(df_p['mae'].values[0])
    
tuning_results = pd.DataFrame(all_params)
tuning_results['mae'] = maes

tuning_results
```

```python
best_params = all_params[np.argmin(maes)]
print(best_params)
```

### 19.5.2 Experiment: Can SARIMA do better? 

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from tqdm import tqdm_notebook
from itertools import product
from typing import Union
```

```python
ad_fuller_result = adfuller(df['y'])

print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```

```python
y_diff = np.diff(df['y'], n=1)

ad_fuller_result = adfuller(y_diff)

print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```

$d=1$, $D=0$ and $m=12$

```python
def optimize_SARIMAX(endog: Union[pd.Series, list], exog: Union[pd.Series, list], order_list: list, d: int, D: int, s: int) -> pd.DataFrame:
    
    results = []
    
    for order in tqdm_notebook(order_list):
        try: 
            model = SARIMAX(
                endog,
                exog,
                order=(order[0], d, order[1]),
                seasonal_order=(order[2], D, order[3], s),
                simple_differencing=False).fit(disp=False)
        except:
            continue
            
        aic = model.aic
        results.append([order, model.aic])
        
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q,P,Q)', 'AIC']
    
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)
    
    return result_df
```

```python
ps = range(0, 4, 1)
qs = range(0, 4, 1)
Ps = range(0, 4, 1)
Qs = range(0, 4, 1)

order_list = list(product(ps, qs, Ps, Qs))

d = 1
D = 0
s = 12
```

```python
SARIMA_result_df = optimize_SARIMAX(train['y'], None, order_list, d, D, s)
SARIMA_result_df
```

```python
SARIMA_model = SARIMAX(train['y'], order=(1,1,1), seasonal_order=(1,0,1,12), simple_differencing=False)
SARIMA_model_fit = SARIMA_model.fit(disp=False)

print(SARIMA_model_fit.summary())
```

```python
SARIMA_model_fit.plot_diagnostics(figsize=(10,8));
```

```python
residuals = SARIMA_model_fit.resid

lbvalue, pvalue = acorr_ljungbox(residuals, np.arange(1, 11, 1))

print(pvalue)
```
